```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

#Empezamos con las condiciones iniciales:

x0 = 0       # posición inicial
v0 = 1       # velocidad inicial
a = 1        # aceleración
t_max = 10   # tiempo total de la animación (s)
dt = 0.1     # intervalo de tiempo entre cuadros (s)


t_vals = np.arange(0, t_max, dt)
x_vals = x0 + v0 * t_vals + 0.5 * a * t_vals**2

fig, eje = plt.subplots()
eje.set_xlim(0, np.max(x_vals) + 5)
eje.set_ylim(-1, 1)
eje.set_xlabel("Posición (m)")
eje.set_title("MRUV - Movimiento Rectilíneo Uniformemente Variado")

point, = eje.plot([], [], 'ro', markersize=10)

camino, = eje.plot([], [], 'b-', lw=2)

# Datos para la trayectoria
x_data = []

def init():
    point.set_data([], [])
    camino.set_data([], [])
    return point, camino

def update(frame):
    x = x_vals[frame]
    x_data.append(x)
    point.set_data(x, 0)
    camino.set_data(x_data, [0]*len(x_data))
    return point, camino

ani = FuncAnimation(fig, update, frames=len(t_vals),
                    init_func=init, blit=True, interval=dt*100)

plt.show()
```

**MOVIMIENTO RECTILÍNEO UNIFORMENTE** \\

- **Ecuación:**  
  $$
    x(t) = x_0 + v\,t
  $$
  donde  
  - $(x_0)$ = posición inicial  
  - $(v)$ = velocidad constante  
  - $(t)$ = tiempo transcurrido  

- **Interpretación:**  
  La partícula se desplaza a velocidad constante, por lo que el espacio recorrido es proporcional a \(t\).


```python
#ejemplo 1 esfera
pelota=sphere(pos=vector(0,0,0), radius=0.5, color=color.blue)

#Definimos  movimiento : MRU

v=vector(1,0,0)  # velocidad

while pelota.pos.x < 10:
  rate(50) # número de iteraciones por segundo
  pelota.pos += v*0.1
  #pelota. pos = pelota.pos + v*0.1
```


**MOVIMIENTO RECTILÍNEO UNIFORMENTE VARIADO** \\

- **Ecuación:**  
  $$
    x(t) = x_0 + v_0\,t + \tfrac{1}{2}\,a\,t^2
  $$
  donde  
  - $(v_0)$ = velocidad inicial  
  - $(a)$ = aceleración constante  

- **Interpretación:**  
  La aceleración convierte el MRU en un movimiento cuyo desplazamiento crece cuadráticamente con el tiempo.



```python
esfera = sphere(pos=vector(0,0,0), radius=0.2, color=color.green)
v = vector(0, 0, 0)   # En reposo
a = vector(0, -9.8, 0) # Aceleración de la gravedad

t = 0
dt = 0.01
while esfera.pos.y > -5:
    rate(10)
    v += a * dt
    esfera.pos += v * dt
    t += dt
```

**MOVIMIENTO CIRCULAR** \\

- **Ecuaciones paramétricas:**  
  $$
    \begin{cases}
      x(t) = R \cos(\omega\,t)\\
      y(t) = R \sin(\omega\,t)\\
      z(t) = v_z\,t
    \end{cases}
  $$
  donde  
  - $(R)$ = radio de la proyección en el plano XY  
  - $(\omega)$ = velocidad angular  
  - $(v_z)$ = velocidad constante en Z  

- **Interpretación:**  
  Movimiento circular en XY con desplazamiento lineal en Z → espiral helicoidal.



```python
#Mostrando la trayectoria
ball = sphere(pos=vector(1,0,0), radius=0.1, color=color.red, make_trail=True)
theta = 0
r = 1

while True:
    rate(100)
    theta += 0.05
    ball.pos = vector(r*cos(theta), r*sin(theta), 0)
```

- **Interpretación:**

**Péndulo Amortiguado**

- **Descripción física:**  
  Un péndulo simple con rozamiento viscous, donde existe un término de amortiguamiento proporcional a la velocidad angular.

- **Ecuación de movimiento (no lineal):**  
  $$
    \ddot\theta(t) + \frac{c}{mL^2}\,\dot\theta(t) + \frac{g}{L}\,\sin\bigl(\theta(t)\bigr) = 0
  $$  
  donde  
  - $(\theta(t))$ = ángulo de desviación  
  - $(c)$ = coeficiente de fricción viscosa  
  - $(m)$ = masa  
  - $(L)$ = longitud del hilo  
  - $(g)$ = aceleración de la gravedad  


```python
from vpython import sphere, curve, vector, rate, color
from math import sin, pi

# Parámetros del sistema
L     = 10         # longitud (m)
m     = 1          # masa (kg)
g     = 9.8        # gravedad (m/s²)
c     = 0.5        # coeficiente de fricción viscosa (kg·m²/s)
dt    = 0.01       # paso de tiempo (s)
theta = pi/4       # ángulo inicial (rad)
dtheta= 0          # velocidad angular inicial (rad/s)

# Configuración visual
base = curve(pos=[vector(-L,0,0), vector(L,0,0)], color=color.white)
pivot= vector(0,0,0)
pos = vector(L*sin(theta), 0, L*cos(theta))
cuerda = curve(pos=[pivot, pos], color=color.yellow)
masa   = sphere(pos=pos, radius=0.3, color=color.red)
```

```python
# Bucle de simulación
while True:
    rate(100)
    # Ecuación de movimiento amortiguado: θ¨ = –(g/L)·sinθ – (c/(mL²))·θ˙
    ddtheta = - (g/L)*sin(theta) - (c/(m*L**2))*dtheta
    dtheta = dtheta + ddtheta*dt
    theta  = theta  + dtheta*dt

    # Normalizamos ángulo para evitar desbordes
    if abs(theta) > 2*pi:
        theta = theta % (2*pi)

    # Actualizamos posición
    pos = vector(L*sin(theta), 0, L*cos(theta))
    cuerda.modify(1, pos=pos)
    masa.pos = pos
```

**Péndulo Forzado y Amortiguado**

- **Descripción física:**  
  Un péndulo simple con rozamiento viscoso **y** una fuerza externa periódica (forzamiento armónico).

- **Ecuación de movimiento (no lineal):**  
  $$
    \ddot\theta + \frac{c}{mL^2}\,\dot\theta + \frac{g}{L}\,\sin\bigl(\theta\bigr)
    = \frac{A}{mL^2}\,\cos\bigl(\Omega\,t\bigr)
  $$
  donde  
  - $(c)$ = coeficiente de rozamiento viscoso  
  - $(A)$ = amplitud del torque forzante  
  - $(\Omega)$ = frecuencia angular de la fuerza externa

```python
from vpython import sphere, curve, vector, rate, color
from math import sin, cos, pi

# Parámetros del sistema
L      = 10        # longitud (m)
m      = 1         # masa (kg)
g      = 9.8       # gravedad (m/s²)
c      = 0.5       # coeficiente de rozamiento (kg·m²/s)
A      = 1.0       # amplitud del torque forzante (N·m)
Omega  = 1.5       # frecuencia de forzamiento (rad/s)
dt     = 0.01      # paso de tiempo (s)

# Condiciones iniciales
theta  = pi/6      # ángulo inicial (rad)
dtheta = 0         # velocidad angular inicial (rad/s)

# Visualización
pivot = vector(0,0,0)
pos   = vector(L*sin(theta), 0, L*cos(theta))
cuerda = curve(pos=[pivot, pos], color=color.cyan)
masa   = sphere(pos=pos, radius=0.3, color=color.magenta)
```

```python
t = 0.0
while True:
    rate(100)
    # Ecuación forzada y amortiguada:
    # θ¨ = –(g/L)·sinθ – (c/(mL²))·θ˙ + (A/(mL²))·cos(Ω·t)
    ddtheta = - (g/L)*sin(theta) \
              - (c/(m*L**2))*dtheta \
              + (A/(m*L**2))*cos(Omega*t)
    dtheta += ddtheta * dt
    theta  += dtheta  * dt
    t      += dt

    # Normalizar ángulo
    theta = theta % (2*pi)

    # Actualizar posición gráfica
    pos = vector(L*sin(theta), 0, L*cos(theta))
    cuerda.modify(1, pos=pos)
    masa.pos = pos
```

**Péndulo Doble (Caótico)**

- **Descripción física:**  
  Dos masas puntuales $(m_1)$ y $(m_2)$ unidas por hilos de longitudes $(L_1)$ y $(L_2)$. El primer péndulo cuelga de un punto fijo y el segundo cuelga del primero. Este sistema muestra comportamiento caótico para ciertas energías iniciales.

- **Ecuaciones de movimiento:**  
  Definimos los ángulos $(\theta_1(t))$ y $(\theta_2(t))$ medidos desde la vertical, con velocidades angulares $(\dot\theta_1)$, $(\dot\theta_2)$. Sea  
  $$
    \Delta = \theta_1 - \theta_2.
  $$  
  Entonces, usando Lagrange, las aceleraciones son:
  $$
  \begin{aligned}
    \ddot\theta_1 &= \frac{
      -g(2m_1 + m_2)\,\sin\theta_1
      - m_2\,g\,\sin(\theta_1 - 2\theta_2)
      - 2\,\sin\Delta\,m_2
        \bigl(\dot\theta_2^2\,L_2 + \dot\theta_1^2\,L_1\cos\Delta\bigr)
    }{
      L_1\bigl(2m_1 + m_2 - m_2\cos(2\Delta)\bigr)
    },\\[6pt]
    \ddot\theta_2 &= \frac{
      2\,\sin\Delta\,
      \bigl(\dot\theta_1^2\,L_1\,(m_1 + m_2)
      + g\,(m_1 + m_2)\cos\theta_1
      + \dot\theta_2^2\,L_2\,m_2\cos\Delta\bigr)
    }{
      L_2\bigl(2m_1 + m_2 - m_2\cos(2\Delta)\bigr)
    }.
  \end{aligned}
  $$

- **Interpretación:**  
  Pequeñas diferencias en las condiciones iniciales conducen a trayectorias muy distintas — característica del caos.


```python
from vpython import sphere, cylinder, vector, rate, color
from math import sin, cos, pi

# Parámetros físicos
m1, m2 = 1.0, 1.0      # masas (kg)
L1, L2 = 5.0, 5.0      # longitudes (m)
g = 9.8                # gravedad (m/s²)
dt = 0.01              # paso de tiempo (s)

# Condiciones iniciales
theta1, theta2 = pi/2, pi/2 + 0.01
dtheta1, dtheta2 = 0.0, 0.0
```

```python
# Elementos gráficos
pivot = vector(0, 0, 0)
bob1 = sphere(pos=vector(L1*sin(theta1), -L1*cos(theta1), 0), radius=0.3, color=color.red)
rod1 = cylinder(pos=pivot, axis=bob1.pos - pivot, radius=0.05, color=color.white)

bob2 = sphere(pos=bob1.pos + vector(L2*sin(theta2), -L2*cos(theta2), 0), radius=0.3, color=color.blue)
rod2 = cylinder(pos=bob1.pos, axis=bob2.pos - bob1.pos, radius=0.05, color=color.white)
```

```python
# Función para actualizar aceleraciones
def accelerations(theta1, theta2, dtheta1, dtheta2):
    dtheta = theta1 - theta2
    denom1 = L1*(2*m1 + m2 - m2*cos(2*dtheta))
    denom2 = L2*(2*m1 + m2 - m2*cos(2*dtheta))
    dd1 = (
        -g*(2*m1 + m2)*sin(theta1)
        - m2*g*sin(theta1 - 2*theta2)
        - 2*sin(dtheta)*m2*(dtheta2**2*L2 + dtheta1**2*L1*cos(dtheta))
    )/denom1
    dd2 = (
        2*sin(dtheta)*(
            dtheta1**2*L1*(m1 + m2)
            + g*(m1 + m2)*cos(theta1)
            + dtheta2**2*L2*m2*cos(dtheta)
        )
    )/denom2
    return dd1, dd2

# Bucle de simulación
while True:
    rate(100)
    # Computar aceleraciones
    dd1, dd2 = accelerations(theta1, theta2, dtheta1, dtheta2)
    # Actualizar velocidades y ángulos (Euler explícito)
    dtheta1 += dd1 * dt
    dtheta2 += dd2 * dt
    theta1  += dtheta1 * dt
    theta2  += dtheta2 * dt

    # Calcular posiciones cartesianas
    pos1 = vector(L1*sin(theta1), -L1*cos(theta1), 0)
    pos2 = pos1 + vector(L2*sin(theta2), -L2*cos(theta2), 0)

    # Actualizar gráficos
    bob1.pos = pos1
    rod1.axis = pos1 - pivot
    bob2.pos = pos2
    rod2.pos  = pos1
    rod2.axis = pos2 - pos1
```

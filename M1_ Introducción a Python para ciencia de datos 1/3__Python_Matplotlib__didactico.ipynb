### Ploteo con leyendas

📏 `np.linspace(inicio, fin, num)` crea un array con valores equidistantes.

$$
y_{i,j} = \sin(x_i + \phi_j)
$$

Es decir, la matriz \( y \in \mathbb{R}^{1000 \times 4} \) contiene las siguientes funciones:

- $ y[:, 0] = \sin(x) $
- $ y[:, 1] = \sin\left(x + \frac{\pi}{2}\right) = \cos(x) $
- $ y[:, 2] = \sin(x + \pi) = -\sin(x) $
- $ y[:, 3] = \sin\left(x + \frac{3\pi}{2}\right) = -\cos(x) $


```python
#nueva data
x = np.linspace(0, 10, 1000)
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
```

📈 `plt.plot(x, y)` genera un gráfico de líneas con los datos x e y.

`plt.legend(frameon=False)`

```python
# 1ro
plt.plot(x, y[:,0], 'r:', label='sin(x)')
# 2do
plt.plot(x, y[:,1], 'b--', label='sin(x + π/2)')

plt.legend()

plt.xlabel('time')
plt.ylabel('sen(x)')
plt.title('Líneas de senos')
plt.show()
```

📈 `plt.plot(x, y)` genera un gráfico de líneas con los datos x e y.

`plt.legend(frameon=True)`

```python
# 1ro
plt.plot(x, y[:,0], 'r:', label='sin(x)')
# 2do
plt.plot(x, y[:,1], 'b--', label='sin(x + π/2)')

plt.legend(frameon=True, loc='lower center')
#'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center

plt.xlabel('time')
plt.ylabel('sen(x)')
plt.title('Líneas de senos')
plt.show()
```

📈 `plt.legend(loc='',bbox_to_anchor=(,))`  leyenda afuera

```python
plt.figure()
plt.plot(x, np.sin(x-0), 'g-.',label='lineal')
plt.plot(x, np.sin(x-1), 'bo-', label='circle')
plt.plot(x, np.sin(x-2), 'r:', label='dots')
plt.plot(x, np.sin(x-3), 'k^-', label='triangle')
plt.legend(loc='upper right',bbox_to_anchor=(1.3,1.1))
plt.xlabel('time', fontsize=20)
plt.title('$sin(x^2)$', fontsize=25)
plt.show()
```

### Fill

📈 `plt.fill_between()` llenar entre gráficos.

```python
# llenado entre gráficas
plt.figure()
plt.plot(x, np.sin(x-0), 'g-.',label='lineal')
plt.plot(x, np.sin(x-3), 'bo-', label='circle')
plt.fill_between(x, np.sin(x-0),color='b')
plt.fill_between(x, np.sin(x-3),color='k')
plt.legend(loc='upper right',bbox_to_anchor=(1.3,1.1))
plt.title('$sin(x^2)$', fontsize=25)
plt.show()
```

## Countour Plots

$$f(x,y)=sen(x)^{10}+cos(10+xy)\times cos(x)$$

```python
#función f
def f(x, y):
    return np.sin(x)**10 + np.cos(10+x*y)*np.cos(x)
```

```python
# nueva data
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

`plt.contour` dibuja líneas de contorno—curvas que unen puntos con un mismo valor—para una función Z=f(X,Y)

```python
# contour + cmap
plt.figure()
plt.contour(X,Y,Z, cmap='inferno')
plt.colorbar()
plt.show()
```

```python
# contourf
plt.figure(figsize=(10,7))
plt.contourf(X,Y,Z, cmap='inferno')
plt.colorbar()
plt.show()
```

### 3D Line Plot

📏 debemos crear la función `Z(x,y)`.

```python
# mplt inicio
fig = plt.figure(figsize=(15, 12))
ax = plt.axes(projection="3d")

# data
z_line = np.linspace(0, 15, 1000)
x_line = np.cos(z_line)
y_line = np.sin(z_line)

# plot line
ax.plot3D(x_line,y_line,z_line)

# labels
ax.set_xlabel('X Axis', color='red', fontsize=16)
ax.set_ylabel('Y Axis', color='red', fontsize=16)
ax.set_zlabel('Z Axis', color='red', fontsize=16)
ax.set_title('line', color='red', fontsize=16);

# angle
ax.view_init(22, 55)

plt.show()
```

# Seaborn



```python
import seaborn as sns
import pandas as pd
```

```python
# Estilo de gráficos por defecto
sns.set(style="whitegrid")
sns.set_palette('pastel')
```

```python
# DataFrame con datos sintéticos
np.random.seed(42)
n = 100
df = pd.DataFrame({
    'x': np.linspace(0, 10, n),
    'grupo': np.random.choice(['A', 'B'], size=n),
    'valor': np.random.normal(loc=0, scale=1, size=n)
})
df['valor'] += df['x'] * (df['grupo'] == 'A')
```

```python
df.head(5)
```

`sns.lineplot()` — Línea con tendencia

```python
plt.figure()
sns.lineplot(data=df, x='x', y='valor', hue='grupo')
plt.title('Tendencia por grupo')
plt.show()
```

 `sns.scatterplot()` — Dispersión con agrupamiento

```python
plt.figure()
sns.scatterplot(data=df, x='x', y='valor', hue='grupo', style='grupo')
plt.title('Gráfico de dispersión por grupo')
plt.show()
```

`sns.histplot()` — Histograma

```python
plt.figure()
sns.histplot(data=df, x='valor', hue='grupo', kde=True, element='step')
plt.title('Distribución de valores')
plt.show()
```

`sns.kdeplot()` — Estimación de densidad de *Kernel*

```python
plt.figure()
sns.kdeplot(data=df, x='valor', hue='grupo', fill=True)
plt.title('Estimación de densidad (KDE)')
plt.show()
```

`sns.pairplot()` — Matriz de gráficos por pares

```python
plt.figure(figsize=(15,15))
sns.pairplot(df, hue='grupo')
plt.suptitle('Relaciones entre variables', y=1.02)
plt.show()
```

`sns.boxplot()` — Diagrama de cajas

```python
plt.figure()
sns.boxplot(data=df, x='grupo', y='valor')
plt.title('Comparación entre grupos')
plt.show()
```

`sns.jointplot()` — Relación bivariada + distribuciones marginales

```python
plt.figure()
sns.jointplot(data=df, x='x', y='valor', hue='grupo', kind='scatter')
#'scatter', 'hist', 'hex', 'kde', 'reg', 'resid'
plt.show()
```

`sns.lmplot()` — Dispersión con regresión lineal

```python
sns.lmplot(data=df, x='x', y='valor')
```

```python
plt.figure()
sns.lmplot(data=df, x='x', y='valor', hue='grupo')
plt.show()
```

`sns.heatmap()` — Mapa de calor

```python
corr = df[['x', 'valor']].corr()

plt.figure()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Mapa de calor de correlaciones')
plt.show()
```

📚 Guía rápida: ¿Qué gráfico usar en Seaborn?

| Tipo de variables       | Objetivo del análisis                                   | Gráfico recomendado           | Función de Seaborn          |
|-------------------------|----------------------------------------------------------|-------------------------------|-----------------------------|
| **1 numérica**          | Ver la distribución de una variable                     | Histograma                    | `sns.histplot()`            |
|                         | Ver una distribución suavizada                          | KDE (densidad)                | `sns.kdeplot()`             |
|                         | Comparar medias por grupo                               | Gráfico de barras             | `sns.barplot()`             |
| **1 categórica, 1 num** | Comparar distribuciones entre grupos                    | Boxplot                       | `sns.boxplot()`             |
|                         | Comparar forma + distribución                           | Violín (violin plot)          | `sns.violinplot()`          |
|                         | Mostrar todos los puntos                                | Stripplot                     | `sns.stripplot()`           |
|                         | Mostrar puntos sin superposición                        | Swarmplot                     | `sns.swarmplot()`           |
| **2 numéricas**         | Relación punto a punto                                  | Dispersión (scatter)          | `sns.scatterplot()`         |
|                         | Relación + línea de regresión                           | Regresión lineal              | `sns.lmplot()`              |
|                         | Ver errores (residuos) de regresión                     | Residual plot                 | `sns.residplot()`           |
|                         | Ver tendencias por grupo o tiempo                       | Línea suavizada               | `sns.lineplot()`            |
|                         | Ver relaciones entre muchas variables numéricas         | Pairplot                      | `sns.pairplot()`            |
|                         | Ver relación + histogramas marginales                   | Jointplot                     | `sns.jointplot()`           |
| **Matriz de variables** | Ver correlaciones o relaciones en una matriz            | Mapa de calor (heatmap)       | `sns.heatmap()`             |

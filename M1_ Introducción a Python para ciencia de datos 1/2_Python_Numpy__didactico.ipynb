### Minkowski Distance Metric

La distancia de Minkowski de orden $(p)$ entre dos vectores $( \mathbf{x}, \mathbf{y} \in \mathbb{R}^n )$ se define como  

$
d_p(\mathbf{x},\mathbf{y}) = \left( \sum_{i=1}^{n} |x_i - y_i|^{\,p} \right)^{1/p},
$

donde $(p > 0)$. Casos particulares:

| \(p\)          | Nombre común | Fórmula simplificada |
|----------------|--------------|----------------------|
| (p = 1)      | Manhattan / L‑1 | $( \sum |x_i-y_i| )$ |
| (p = 2)      | Euclidiana / L‑2 | $( \sqrt{\sum (x_i-y_i)^2} )$ |
| (p -> ∞) | Chebyshev / L‑∞ | $( \max |x_i-y_i| )$ |

---

Implementación con NumPy


def minkowski_distance(x, y, p=2):
    """
    Calcula la distancia de Minkowski entre dos vectores 1‑D usando NumPy.
    - x, y: array‑like, misma longitud.
    - p: >0 o np.inf.
    Devuelve: float.
    """
    # 1️⃣ Convierte los argumentos a arrays de tipo float
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)
    
    # 2️⃣ Verifica que tengan la misma forma
    if x.shape != y.shape:
        raise ValueError("Los vectores deben tener la misma forma.")
    if p <= 0:
        raise ValueError("El parámetro 'p' debe ser positivo.")
    
    # 3️⃣ Caso especial Chebyshev (p = ∞)
    if p == np.inf or p == float("inf"):
        return np.max(np.abs(x - y))
    
    # 4️⃣ Fórmula general
    return np.linalg.norm(x - y, ord=p)


```python
def minkowski_distance(x, y, p=2):
    """
    Calcula la distancia de Minkowski entre dos vectores 1‑D usando NumPy.
    - x, y: array‑like, misma longitud.
    - p: >0 o np.inf.
    Devuelve: float.
    """
    # 1️⃣ Convierte los argumentos a arrays de tipo float
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)

    # 2️⃣ Verifica que tengan la misma forma
    if x.shape != y.shape:
        raise ValueError("Los vectores deben tener la misma forma.")
    if p <= 0:
        raise ValueError("El parámetro 'p' debe ser positivo.")

    # 3️⃣ Caso especial Chebyshev (p = ∞)
    if p == np.inf or p == float("inf"):
        return np.max(np.abs(x - y))

    return np.linalg.norm(x - y, ord=p)
```

```python
if __name__ == "__main__":
    print("Euclidiana (p=2)  :", minkowski_distance([0, 0], [3, 4], p=2))     # 5.0
    print("Manhattan  (p=1)  :", minkowski_distance([0, 0], [3, 4], p=1))     # 7.0
    print("Chebyshev (p=inf):", minkowski_distance([0, 0], [3, 4], p=np.inf)) # 4.0
```
